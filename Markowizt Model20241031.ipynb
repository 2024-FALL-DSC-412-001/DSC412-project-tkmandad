```python
# Downloading data for all tickers at once
data = yf.download(tickers, start='2014-10-09', end='2024-10-09')['Adj Close']
data.index = data.index.date
data
```

```python
df_log_returns = np.log(data / data.shift(1)).dropna()
mu_log = df_log_returns.mean()
```

```python
def subsets(tickers: list[str]):
    def backtrack(curr, i):
        if i > len(tickers):
            return

        ans.append(curr[:])
        for j in range(i, len(tickers)):
            curr.append(tickers[j])
            backtrack(curr, j + 1)
            curr.pop()

    ans = []
    backtrack([], 0)
    return ans
```

```python
def efficient_portfolio(data):

    # Covariance matrix for this portfolio
    C = data.cov()
    C_inv = np.linalg.inv(C)
    
    ones = np.ones(len(C))
    new_weights = np.dot(C_inv, ones) / np.dot(ones.T, np.dot(C_inv, ones))
    
    daily_log_return = mu_log[data.columns].dot(new_weights)
    total_log_return = daily_log_return * 250
    variance = np.dot(new_weights.T, np.dot(C * 250, new_weights))

    return [total_log_return, variance]
```

```python
portfolios = []
n = 10000
df=pd.DataFrame(data={'return':np.zeros(n),'std':np.zeros(n)})

for i in range(n):
    assets = random.sample(tickers, 10)
    input = pd.DataFrame(df_log_returns[assets])
    output = efficient_portfolio(input)
    
    portfolios.append(assets)
    df.loc[i, 'return'] = output[0]
    df.loc[i, 'std'] = output[1] ** 0.5
```

```python
df.plot(x = 'std',y = 'return', kind = 'scatter', figsize = (15,5))
plt.xlabel('Expected risk')
plt.ylabel('Expected Return')
plt.show()
```

```python
# Download S&P 500 data from Yahoo Finance and prepare the benchmark data
sp500_data = yf.download('^GSPC', start='2014-10-09', end='2024-10-09') 
sp500_prices = sp500_data['Adj Close']
sp500_log_returns = np.log(sp500_prices / sp500_prices.shift(1)).dropna() 

sp500_annual_log_return = sp500_log_returns.mean() * 250
sp500_std = sp500_log_returns.std() * (250 ** 0.5)
```

```python
df.plot(x = 'std',y = 'return', kind = 'scatter', figsize = (15,5))
plt.xlabel('Expected risk')
plt.ylabel('Expected Return')

plt.scatter(sp500_std, sp500_annual_log_return, color='red', marker='o', s=60)
plt.axhline(y=sp500_annual_log_return, color='blue', linestyle='--')
plt.axvline(x=sp500_std, color='blue', linestyle='--')

plt.show()
```
